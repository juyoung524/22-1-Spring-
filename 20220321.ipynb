### Constructing a DataFrame using dict

Dict를 활용해서 [  ] (준괄호), 안에 들어가는 내용들이 구분되어 있음

Dict 안에 들어가는 내용들 총 5개, comma로 분리되어 있음

각 아이템들은 앞부분에 있는 key와 column을 이용해서 key와 value로 구분됨

총 5개 정의, 각 item에 대해 :로 왼쪽, 오른쪽으로 입력 가능

위에도 다섯 개니까 밑에도 다섯 개, 개수가 맞아야 함

```python
# Some input data - a small sample of the iris dataset
data = {'sepal_length': [6.9, 6.9, 4.8, 5.4, 4.6],
        'sepal_width': [3.2, 3.1, 3.4, 3.0, 3.6],
        'petal_length': [5.7, 5.1, 1.9, 4.5, 1.0],
        'petal_width': [2.3, 2.3, 0.2, 1.5, 0.2],
        'species': ['virginica', 'virginica', 'setosa', 'versicolor', 'setosa']}

df = pd.DataFrame(data)
#Dict를 넣어주면 dataframe으로 만들어 줌
df 
```

Python 내부에서 dict 변수를 직접 하나하나 적어서 DataFrame이라는 function 통해서 dataframe 변수 만들어줄 수도 있음

### Constructing a DataFrame using pd.Series

한 column의 내용에 해당하는 내용 따로 만드는 것임

sepal_length

sepal_width...

해서 총 5개

내용에 해당하는 부분만 5개를 pandas에 들어있는 'series'라는 함수로 만들어줌

```python
sepal_length = pd.Series([6.9, 6.9, 4.8, 5.4, 4.6])
sepal_width = pd.Series([3.2, 3.1, 3.4, 3.0, 3.6])
petal_length = pd.Series([5.7, 5.1, 1.9, 4.5, 1.0])
petal_width = pd.Series([2.3, 2.3, 0.2, 1.5, 0.2])
species = pd.Series(['virginica', 'virginica', 'setosa', 'versicolor', 'setosa'])

df = pd.DataFrame({'sepal_length': sepal_length, 'sepal_width': sepal_width, 'petal_length': petal_length, 'petal_width': petal_width, 'SPECIES': species})
df

#Dict 여기서 만들어줌. 
```

1) CSV 파일 불러와서 만드는 것
-> 주로 이용하는 방법

2) Dict 파일을 통으로 다 만들어서 만드는 방법

3) Series라는 것 써서 칼럼의 내용 부분만 따로 만들고 밑에서 데이터프레임 만드는 것

### Changing column names

Column에 대한 정보 저런 식으로 나옴.

```python
# Change the columns
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
df

#총 5개: species 다시 assign함
#실행하면 species 부분에 column명이 소문자로 바뀌어 있음
```

### Accessing columns

```python
# Access one column 
df.loc[:, 'sepal_length']

#comma 왼쪽은 row, 오른쪽은 column의미함
#:는 '전부 다'라는 것 의미 (행에 대해 선부 다라는 것을 의미함)
#sepal length에 대한 정보만 다 가져옴 (column에서)
```

```python
# Access one column (short-hand)
df['sepal_length']
```

df[column 명] 이렇게만 적어도 더 간단히 쓸 수 있음

```python
# Access one column (shortest-hand)
df.sepal_length
```

df.sepal_length 이렇게만 해도 더 간단히 정보 가져올 수 있음

```python
# Access two columns
df.loc[:, ['sepal_length', 'sepal_width']]
```

두 개의 정보 다 가져올 수 있음 (comma하고 여러 개 쓰면)

```python
# Access two columns (short-hand)
df[['sepal_length', 'sepal_width']]
```

### Accessing rows

```python
# Access one row 
df.loc[3, :]
```

### Adding/Removing new columns

```python
df['new_column'] = 1
df
```

column access하는 쉬운 방법 중 하나임

Column이 새로 생겨버림 (new_column) 1을 다 assign한다는 의미임 (새로운 column 생기는 것)

```python
df['new_column'] = [1, 2, 3, 4, 5]
df
```

1, 2, 3, 4, 5로 바뀜

```python
df = df.drop('new_column', axis=1) # axis=1: column
df
```

axis=1: new column는 axis 1에서 찾아라!

0이 먼저, 1이 두 번째임

drop시키고 원래 있던 df로 해야지 업데이트됨

### Adding/Removing new rows

```python
df.loc[5, :] = ''
df
```

row 4까지밖에 없는데, 저렇게 적고 '  ' 적으면 (string) string 안에 아무것도 없는 blank string을 다섯 번째 row 만들어거 그걸 다 넣으라는 말임

실행하면 다섯 번째 row 만들어지고, blank로 비어 있도록 실행됨

```python
df = df.drop(5, axis=0) # axis=0: row
df
```

axis 0하면 row를 찾는 것임

다섯 번째 행축을 찾아서 drop하고 df로 assign하라
