챗봇 모델 중에 제일 괜찮은 모델임

```python
from transformers import Conversation
Conversation("Going to the movies tonight - any suggestions?")

conversation_1 = Conversation("Going to the movies tonight - any suggestions?")
conversation_2 = Conversation("What's the last book you have read?")

pipe([conversation_1, conversation_2])
```

두 개의 문장을 conversation 1, 2로 list해서 pipe에 넣은 것임

Bot이 왔다갔다 (챗봇) 자동적으로 만들어 주는 것임

user이 '나 자신'이고, bot가 '챗봇'임. 첫 번째 내가 오늘 영화보러 갈 건데 뭐 보러 갈까? 하니까 BOt이 어떤 영화 볼지 알려줌.

어떤 책 읽었냐고 물어보니까 답변하고 있음.

```python
conversation_1.add_user_input("Is it an action movie?")
conversation_2.add_user_input("What is the genre of this book?")

pipe([conversation_1, conversation_2])
```

zero shot: 훈련을 할 때 **다 label을 붙임**. 사진 찍고 **이 사진은 '강아지', '고양이'라고 다 가르쳐 줌 (모델한테)**

model한테 한 번도 가르쳐주지 않았음에도 불구하고 무리한 질문들을 해봄. 

```python
pipe("one day I will see the world", ['travel', 'cooking', 'dancing'])
```

앞선 문장이 어떤 유형의 text인지, 세 개에 대한 확률을 줘라. 

Travel 관련된 것이다. 컴퓨터가 이걸 알기가 쉽지 않음 (Dancing도, Cooking도 아님)

여행 관련 말이라는 것을 알아냄

```python
pipe = pipeline(task="zero-shot-image-classification")
```

```python
url = "http://images.cocodataset.org/val2017/000000039769.jpg"
pipe(url, candidate_labels=["a photo of a cat", "a photo of a dog"])
```

네이버나 구글 렌즈로 사진 찍고 이게 뭔지 알아맞춰라 하는 것 (그런 데이터를 미리미리 다 훈련시켜서 할 수 있는 것임)

Zero shot은 훈련 없이 (image classification)하는 거

Candidate labels라고 하고 photo of a dog, cat, monkey라고 적고 실행하면 확률이 나옴. Cat일 확률이 99%라고 도출해 주는 것임.
